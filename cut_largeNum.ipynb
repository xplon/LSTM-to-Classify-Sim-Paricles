```python
# This file is used to remove all the extreme and unusual events

import numpy as np
import pandas as pd

file_path = './data/train/full-fixed_2.parquet'
data = pd.read_parquet(file_path)

df = pd.DataFrame()

columns_to_check = [f'DomId_{i}' for i in range(21)]
# remove columns in which have any intensities in one Dom greater than or equal to 2000
filtered_data = data[~data[columns_to_check].ge(2000).any(axis=1)]

columns_to_check = [f'NDomId_{i}' for i in range(21)]
# remove columns in which have any counts in one Dom greater than or equal to 5000
filtered_data = filtered_data[~filtered_data[columns_to_check].ge(5000).any(axis=1)]

# ang_data.replace(np.nan, -10000, inplace=True)

# filtered_data[columns_to_check] = filtered_data[columns_to_check].replace(-10000, 0)
# filtered_data[columns_to_check] = filtered_data[columns_to_check].fillna(0)

# This file will be processed into train_3.parquet and test_3.parquet
filtered_data.to_parquet('./data/train/full-fixed_3.parquet')
```
